```python
# ANOVA
df_anova=df[["make","price"]] # select the columns on which ANOVA has to be performed
```

```python
grouped_anova=df_anova.groupby(["make"]) # group by 'make' column 
```

```python
import scipy.stats as stats # import scipy.stats as it contains one way anova functionality
```

```python
anova_results_1=stats.f_oneway(grouped_anova.get_group("honda")["price"],grouped_anova.get_group("subaru")["price"])
# use stats.f_oneway to find the f and p-value b/w groups (here-2 groups) to determine weak/strong correlation 
# honda and subaru groups compared
```

```python
print(anova_results_1)  # weak correlation as f value < 1 and p value > 0.05 also
```

```python
anova_results_2=stats.f_oneway(grouped_anova.get_group("honda")["price"],grouped_anova.get_group("jaguar")["price"])
```

```python
print(anova_results_2) # strong correlation as f value > 1 and p value < 0.05 
```
