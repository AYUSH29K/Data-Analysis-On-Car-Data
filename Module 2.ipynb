```python
import pandas as pd
```

```python
import numpy as np 
```

```python
url = "https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/auto.csv"
```

```python
df=pd.read_csv(url)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df.columns=headers
```

```python
df.head(5)
```

```python
df["symboling"]= df["symboling"] + 1 # adding values to the columns (here 1 added to symboling column)
```

```python
df.head(5)
```

```python
df.dropna(subset = ["normalized-losses"], axis = 0 , inplace = True) 
# deleting row (axis =0) deleting column (axis=1) for missing vals
```

```python
mean = df["symboling"].mean() # take mean for missing numerical values and then replace missing vals with the mean
# mode taken for categorical variables
```

```python
df["symboling"].replace(3,mean) 
# after mean is taken replace the missing vals with the mean (here 3 replace with mean of the symboling column)
```

```python
df["city-mpg"]= 235/df["city-mpg"]# to make calculations in all values of a column (here mpg changed to mp-L of city)
```

```python
df.head(5) # sucessfully converted and performed calculation on the city mpg column
```

```python
df.rename(columns={"city-mpg":"city-mp-L"},inplace=True) 
# rename column where calculation done appropriatley(inplace=True)
```

```python
df.head(5)# display to check
```

```python
df["wheel-base"]=df["wheel-base"]/df["wheel-base"].max() 
# normalizing a column by simple feature scaling method (here,wheel base column)
```

```python
df.head(5) # check results
```

```python
df["engine-size"]=(df["engine-size"]-df["engine-size"].min())/(df["engine-size"].max()-df["engine-size"].min())
# normalizing a column using min-max method i.e. xnew=xold-xmin/xmax-xmin (here column is engine-size)
```

```python
df.head(10) #check it!!
```

```python
df["symboling"]=(df["symboling"]-df["symboling"].mean())/df["symboling"].std() 
#normalizing the column with z score method i.e. xnew=x-mu/sigma (here column normalized is symboling)
```

```python
df.head(5) #success!!
```

```python
df.price = pd.to_numeric(df.price, errors='coerce').fillna(0).astype(np.int64) 
# converted price column to int by this method as was having problems
```

```python
bins = np.linspace(min(df["price"]), max(df["price"]), 4)
# binning (grouping of values here price into 3 categories) the column using 4 dividers using linspace function
```

```python
group_names = ["Low","Medium","High"] # divinding the grouped values into categories respectively
```

```python
df["price-binned"] = pd.cut(df["price"],bins,labels=group_names,include_lowest=True) 
# after grouping,cutting into segments and sorting data values into bins using pandas' cut function
```

```python
df.head() # Check the new column price-binned categories have been created i.e. low,medium,high for the column
```

```python
pd.get_dummies(df['fuel-type']) 
#To convert a columns categorical to numerical values first convert them by assigning them dummy variables (0/1)
#(here coln fuel-type)
```

```python
df.head(15) # check here that the column has been assigned dummy vars (here 0 for one feature 1 for other feature)
```

```python
dummy_variable_1=pd.get_dummies(df['body-style']) 
#example for more than 2 variables assign dummy_variable_1 to dummy vars created after assinging them to body style
```

```python
df = pd.concat([df, dummy_variable_1], axis=1) #merge df to dummy_variable_1
```

```python
df.drop("body-style", axis = 1, inplace=True) 
#drop the column (originally body style) as replace with dummy vars no of each category (0/1)
```
